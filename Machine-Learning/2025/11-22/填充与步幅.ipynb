填充

```python
import torch
from torch import nn
# 定义⼀一个函数来计算卷积层。它对输⼊入和输出做相应的升维和降维
def comp_conv2d(conv2d, X): 
    # (1, 1)代表批量量⼤大⼩小和通道数（“多输⼊入通道和多输出通道”⼀一节将介绍）均为1
    X = X.view((1, 1) + X.shape)
    Y = conv2d(X)
    return Y.view(Y.shape[2:]) # 排除不不关⼼心的前两维：批量量和通道
# 注意这⾥里里是两侧分别填充1⾏行行或列列，所以在两侧⼀一共填充2⾏行行或列列
conv2d = nn.Conv2d(in_channels=1, out_channels=1, kernel_size=3,padding=1)
X = torch.rand(8, 8)
comp_conv2d(conv2d, X).shape

# 使⽤用⾼高为5、宽为3的卷积核。在⾼高和宽两侧的填充数分别为2和1
conv2d = nn.Conv2d(in_channels=1, out_channels=1, kernel_size=(5,3), padding=(2, 1))
comp_conv2d(conv2d, X).shape
```

步幅

```python
conv2d = nn.Conv2d(1, 1, kernel_size=3, padding=1, stride=2)
comp_conv2d(conv2d, X).shape

conv2d = nn.Conv2d(1, 1, kernel_size=(3, 5), padding=(0, 1), stride=
(3, 4))
comp_conv2d(conv2d, X).shape
```
