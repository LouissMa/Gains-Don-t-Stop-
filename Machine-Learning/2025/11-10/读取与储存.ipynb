1.读写 TENSOR

```python
import torch
from torch import nn

# 创建Tensor并保存
x = torch.ones(3)
torch.save(x, 'x.pt')

# 从文件加载Tensor
x2 = torch.load('x.pt')
print(x2)  # 输出: tensor([1., 1., 1.])
print(torch.equal(x, x2))  # 输出: True

# 保存Tensor列表
y = torch.zeros(4)
torch.save([x, y], 'xy_list.pt')

# 加载Tensor列表
xy_list = torch.load('xy_list.pt')
print(xy_list)  # 输出: [tensor([1., 1., 1.]), tensor([0., 0., 0., 0.])]
print(f"列表长度: {len(xy_list)}")  # 输出: 列表长度: 2

# 保存Tensor字典
torch.save({'x': x, 'y': y}, 'xy_dict.pt')

# 加载Tensor字典
xy_dict = torch.load('xy_dict.pt')
print(xy_dict)  # 输出: {'x': tensor([1., 1., 1.]), 'y': tensor([0., 0., 0., 0.])}
print(f"x: {xy_dict['x']}")  # 输出: x: tensor([1., 1., 1.])
print(f"y: {xy_dict['y']}")  # 输出: y: tensor([0., 0., 0., 0.])
```

2.读写模型
    （1）state_dict
    （2）优化器state_dict

```python
import torch
import torch.nn as nn

class MLP(nn.Module):
    def __init__(self):
        super(MLP, self).__init__()
        self.hidden = nn.Linear(3, 2)  # 可学习参数
        self.act = nn.ReLU()           # 无参数
        self.output = nn.Linear(2, 1)  # 可学习参数
    
    def forward(self, x):
        a = self.act(self.hidden(x))
        return self.output(a)

net = MLP()
net.state_dict()  # 获取模型参数字典

```

```python
#优化器state_dict()
optimizer = torch.optim.SGD(net.parameters(), lr=0.001, momentum=0.9)
optimizer.state_dict()
```
